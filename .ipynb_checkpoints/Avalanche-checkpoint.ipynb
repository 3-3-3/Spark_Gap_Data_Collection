```python
def av_frames(V, threshold=9):
    frames = []
    dV = np.diff(V)

    index = 0
    points = 100
    delta_V = 0
    i = 0


    while i < len(dV):
        #print(type(delta_V))
        if np.sign(dV[i]) == 1:
            delta_V += dV[i]

        else:
            if abs(delta_V) > threshold:
                frames.append(V[index : index+points])
                i += (points - 1)


            delta_V = 0
            index = i

        i += 1
    return frames
```

```python
f = av_frames(t,V)
```

```python
f
```

```python
Dt = (t[1] - t[0]) * 1e9
Dt
```

```python
fig, ax = plt.subplots()
t_pulse = np.linspace(0, Dt * 100, 100)

for frame in f:
    ax.plot(t_pulse,frame,marker = '.')
    
ax.set_title('Pulses')
ax.set_xlabel('Time From Beginning of Pulse (ns)')
ax.set_ylabel(r'$V_{out}$ (V)')
```
